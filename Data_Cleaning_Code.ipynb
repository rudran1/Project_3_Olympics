```python
import datetime as dt
```

```python

def ms_conversion(number_in_ms):
    if ":" in number_in_ms:
        return number_in_ms
    elif "est" in number_in_ms:
        return number_in_ms.replace("est", "").strip()
    else:
        seconds = dt.timedelta(milliseconds=int(number_in_ms)).total_seconds()
        hours = seconds//3600
        minutes = (seconds % 3600)//60
        sec = seconds % 60 
        return "{:02d}:{:02d}:{:.3f}".format(int(hours),int(minutes),round(sec,3))

df_results_clean['value_unit'] = df_results_clean['value_unit'].map(ms_conversion)
df_results_clean['value_unit']
```

```python
Path = 'Project_3\Olympic_Data_Paris\2024 Paris Medals.xlsx'

df = pd.read_csv("Project_3/Olympic_Data_Paris/2024 Paris Medals (3).csv")
df_results_p = (df
      .clean_names()  # Standardize column names
      .drop_duplicates()  # Remove duplicate rows
      .fillna(method='ffill')  # Fill missing values
     )
df_results_clean_p = df_results_p[["discipline_title", 'event_title',"slug_game","medal_type", "participant_title", "value_unit"]]
df_results_clean_p.rename(columns={"participant_title": 'country_name'}, inplace=True)
df_results_clean_p['value_unit'] = df_results_clean_p['value_unit'].astype("str")
df_results_clean_p['value_unit'] = df_results_clean_p['value_unit'].map(ms_conversion)
df_results_clean_p['value_unit']
print(df_results_clean_p.head())
```
