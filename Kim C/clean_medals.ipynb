```python
import pandas as pd

medals_path = "Resources/olympic_medals.csv"
ex_teams_path = "Olympic_Data_Paris/Teams.xlsx"
teams_path = "Resources/Teams.csv"
ex_paris_medals = "Resources/2024 Paris Medals.xlsx"
paris_path = "Resources/Paris_Medals.csv"

pm = pd.read_excel(ex_paris_medals)
df = pd.read_excel(ex_teams_path)
csv_file_path = "Resources/Teams.csv"
paris_csv = "Resources/Paris_Medals.csv"

df.to_csv(csv_file_path, index=False)
pm.to_csv(paris_csv, index=False)
```

```python
teams_df = pd.read_csv(teams_path)
medals_df = pd.read_csv(medals_path)
paris_df = pd.read_csv(paris_path)
```

```python
teams_df.rename(columns={
    "Name": "Country",
}, inplace=True)

teams_df['Country-Year'] = 'paris-2024'

teams_df.head()
```

```python
paris_df.rename(columns={
    "discipline_title": "Discipline",
    "slug_game": "Country-Year",
    "event_title": "Event",
    "event_gender": "Gender",
    "medal_type": "Medal Type",
    "country_name": "Country"
}, inplace=True)

paris_df.head()
```

```python
paris_columns = ["Discipline", "Country-Year", "Event", "Medal Type", "Country"]
paris_filtered = paris_df[paris_columns]
paris_filtered.head()
```

```python
medals_df.head()
```

```python
medal_columns = ["discipline_title", "slug_game", "event_title", "event_gender", "medal_type", "country_name"]
medals_filtered = medals_df[medal_columns]
medals_filtered.head()
```

```python
medals_filtered.rename(columns={
    "discipline_title": "Discipline",
    "slug_game": "Country-Year",
    "event_title": "Event",
    "event_gender": "Gender",
    "medal_type": "Medal Type",
    "country_name": "Country"
}, inplace=True)

medals_filtered.head()
```

```python
merged_df = pd.merge(medals_filtered, teams_df, 
                     left_on=['Country', 'Discipline', 'Event', 'Country-Year'], 
                     right_on=['Country', 'Discipline', 'Event', 'Country-Year'], 
                     how='outer')

merged_df.to_csv("Resources/merged_data.csv", index=False)
merged_df.head()
```

```python
merged_columns = ["Discipline", "Country-Year", "Event", "Medal Type", "Country"]
merged_filtered = merged_df[merged_columns]
merged_filtered.head()
# medal_columns = ["discipline_title", "slug_game", "event_title", "event_gender", "medal_type", "country_name"]
# medals_filtered = medals_df[medal_columns]
```

```python
merged_filtered = merged_filtered[['Discipline', 'Country', 'Event', 'Medal Type', 'Country-Year']]
merged_filtered.head()
```

```python
events_sort = [
    "Men's 4 x 400m Relay", "Men's 4 x 100m Relay", "Women's 4 x 400m Relay",
    "Women's 4 x 100m Relay", "Men's 4 x 200m Freestyle Relay", 
    "Men's 4 x 100m Freestyle Relay", "Women's 4 x 100m Freestyle Relay", 
    "Women's 4 x 200m Freestyle Relay"
]

teams_filtered = merged_filtered[merged_filtered['Event'].isin(events_sort)]

sorted_df = teams_filtered.sort_values(by='Event')

sorted_df.to_csv("Resources/event_sorted.csv", index= False)
sorted_df.head(50)
```
